```python
q1=df['education_expenditure'].quantile(0.25)
```

```python
q3=df['education_expenditure'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
```

```python
upper=q3+1.5*IQR
```

```python
df=df[(df['education_expenditure']>lower) & (df['education_expenditure']<upper)]
```

# Top 10 Country with Highest Average Income 

```python
new_df=df[['country','avg_income']]
```

```python
new_df
```

```python
new_df=new_df.sort_values(by="avg_income",ascending=False)[:10]
```

```python
new_df
```

```python
sns.barplot(x=new_df['country'],y=new_df['avg_income'])
plt.xticks(rotation=90)
plt.show()
```

```python
label_df=new_df.groupby("country")['avg_income'].mean().sort_values(ascending=False).index
```

```python
value_df=new_df.groupby("country")['avg_income'].mean().sort_values(ascending=False).values
```

```python
plt.pie(x=value_df,labels=label_df,autopct="%1.2f")
plt.show()
```
